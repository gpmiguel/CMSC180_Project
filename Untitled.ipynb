```python
import pandas as pd

import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
import numpy as np
```

```python
bubbleSortData = pd.read_csv("./Bubble Sort/Serial/bubbleSort.csv") 
pBubbleSortData = pd.read_csv("./Bubble Sort/Parallel/pBubbleSort.csv") 

bucketSortData = pd.read_csv("./Bucket Sort/Serial/bucketSort.csv") 
pBucketSortData = pd.read_csv("./Bucket Sort/Parallel/pBucketSort.csv") 

mergeSortData = pd.read_csv("./Merge Sort/Serial/mergeSort.csv") 
pMergeSortData = pd.read_csv("./Merge Sort/Parallel/pMergeSort.csv") 

shellSortData = pd.read_csv("./Shell Sort/Serial/shellSort.csv") 
pShellSortData = pd.read_csv("./Shell Sort/Parallel/pShellSort.csv") 
```

```python
print(bubbleSortData)

x = bubbleSortData.N
y = bubbleSortData.theo_seconds
plt.plot(x, y)

plt.title("Theoretical Bubble Sort")
plt.xlabel("Sizes of N")
plt.ylabel("Running Time (seconds)")

#plt.savefig("theoBubbleSort.png")
plt.show()
```

```python
print(pBubbleSortData)

x = pBubbleSortData.N
y = pBubbleSortData.seconds
plt.scatter(x, y)

plt.title("Theoretical Parallel Bubble Sort")
plt.xlabel("Sizes of N")
plt.ylabel("Running Time (seconds)")

plt.savefig("pBubbleSort.png")
plt.show()
```
